```python
from IPython.utils.io import temp_pyfile
import json
import pandas as pd

def open_json_to_DF(path):
    with open(path, 'r') as file:
        data = json.load(file)
        data = dict(data)

        fname = data['FileInfo']['FileName']
        if '정상' in fname:
            label = 'normal'
        elif '졸음' in fname:
            label = 'sleep'
        elif '하품' in fname:
            label = 'yawn'
        elif '통화' in fname:
            label = 'call'
        elif '흡연' in fname:
            label = 'smoking'

        d = {
            'FileName' : data['FileInfo']['FileName'],
             'Width': data['FileInfo']['Width'],
             'Height':data['FileInfo']['Height'],
             'label': label,
             'Mask' :data['Accessory']['Mask'],
             'Glasses' : data['Accessory']['Glasses'],
             'Cap': data['Accessory']['Cap'],
             'Annotation' : data['Annotation'],
             'Face_TF': data['ObjectInfo']['BoundingBox']['Face']['isVisible'],
             'Face_BB': str(data['ObjectInfo']['BoundingBox']['Face']['Position']),
             'Leye_TF': data['ObjectInfo']['BoundingBox']['Leye']['isVisible'],
             'Leye_BB': str((data['ObjectInfo']['BoundingBox']['Leye']['Position'])),
             'Reye_TF': data['ObjectInfo']['BoundingBox']['Reye']['isVisible'],
             'Reye_BB': str((data['ObjectInfo']['BoundingBox']['Reye']['Position'])),
             'Mouth_TF': data['ObjectInfo']['BoundingBox']['Mouth']['isVisible'],
             'Mouth_BB': str((data['ObjectInfo']['BoundingBox']['Mouth']['Position'])),
             'Cigar_TF': data['ObjectInfo']['BoundingBox']['Cigar']['isVisible'],
             'Cigar_BB': str((data['ObjectInfo']['BoundingBox']['Cigar']['Position'])),
             'Phone_TF': data['ObjectInfo']['BoundingBox']['Phone']['isVisible'],
             'Phone_BB': str((data['ObjectInfo']['BoundingBox']['Phone']['Position']))
        }

    df = pd.DataFrame(d, index = [0])
    return df
```

```python
fnames = '/content/drive/MyDrive/ZB/DL_project_2/data/Validation/label_val/045_G1/045_G1_01_무광원_계기판_정상주시_20200929_110850_02193.json'

data = open_json_to_DF(fnames)
data
```

```python
base_path = '/content/drive/MyDrive/ZB/DL_project_2/data/Validation/label_val'
folderList = os.listdir(base_path)

df_merge = data

for folder in folderList:
    path = os.path.join(base_path, folder)
    fnames = os.listdir(path)

    for fname in fnames:
        src = os.path.join(path, fname)
        try:
            df = open_json_to_DF(src)
            df_merge = pd.concat([df_merge, df])
        except Exception as e:
            print(e)
            print(fname)
```
